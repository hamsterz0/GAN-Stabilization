```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_data(filename):
    i = 0
    iteration = []
    d_loss = []
    g_loss = []
    with open(filename, 'r') as f:
        for line in f.readlines():
            if line.startswith("*"):
                continue
            data = line.strip().split()
            if data[0] == 'Time':
                continue
            i = i + 1
            iteration.append(i)
            d_loss.append(float(data[2]))
            g_loss.append(float(data[4]))
    return iteration, d_loss, g_loss
```

```python
def plot_graph(filename):
    iteration, d_loss, g_loss = get_data(filename)
    plt.figure(figsize=(10,5))
    plt.title("Generator and Discriminator Loss During Training")
    plt.plot(iteration[::50],d_loss[::50],label="G")
    plt.plot(iteration[::50],g_loss[::50],label="D")
    plt.xlabel("iterations")
    plt.ylabel("Loss")
    plt.legend()
    plt.show()
```

```python
plot_graph('svd_reparam.log')
```
